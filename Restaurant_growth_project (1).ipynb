```python
df["TotalRevenue"]=(df["InStoreRevenue"]+df["UberEatsRevenue"]+df["DoorDashRevenue"]+df["SelfDeliveryRevenue"])
```

```python
df.sort_values("TotalRevenue", ascending=False).head(10)
```

```python
df["TotalProfit"]=(df["InStoreNetProfit"]+df["UberEatsNetProfit"])
```

```python
df.sort_values("TotalProfit", ascending=False).head(10)
```

```python
import numpy as np
```

```python
conditions = [(df["GrowthFactor"]>=80), (df["GrowthFactor"]>=50) & (df["GrowthFactor"]<80), (df["GrowthFactor"]<50)]
```

```python
categories = ["High Growth", "Medium Growth", "Low Growth"]
```

```python
df["GrowthCategory"]=np.select(conditions,categories,default="Low Growth")
```

```python
df["GrowthCategory"].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
df["GrowthCategory"].value_counts().plot(kind="bar")
plt.show()
```

```python
plt.hist(df["TotalRevenue"])
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Number of restaurants")
plt.show()
```

```python
plt.hist(df["MonthlyOrders"])
plt.title("Monthly Orders Distribution")
plt.xlabel("Orders")
plt.ylabel("Number of restaurants")
plt.show()
```
